```python
#Imported pandas to use DataFrame
#Imported matplotlib and seaborn for visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Importing World bank data
file ='WDIW Dataset.xlsx'
wbd=pd.read_excel(file, index_col = 1)
#Filtering the dataset with jetfire
jetfire=wbd[:][wbd['Cool Name']== 'Jetfire']

jetfire
```

```python
#Checking null values in each column
jetfire.isnull().sum()
```

```python
#Plotting a heatmap with Jetfire Dataset
jetfire_corr = jetfire.corr(method ='kendall')
sns.palplot(sns.color_palette('coolwarm', 12))

# specifying plot size
fig, ax = plt.subplots(figsize=(25,25))
# creating a heatmap
sns.heatmap(data       = jetfire_corr,
            cmap       = 'coolwarm',
            square     = True,
            annot      = True,
            linecolor  = 'black',
            linewidths = 0.5)
plt.savefig("heatmap.png")
```

```python
#Checking the percentage of null values in each column
jetfire.isnull().sum()/len(jetfire)
```

```python
#Deleting columns which has only null values
for col in jetfire:
    
    if jetfire[col].isnull().sum()/len(jetfire)==1:
        del jetfire[col]
```
